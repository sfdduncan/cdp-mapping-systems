```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

df = pd.read_csv("./geocoded_by_geoapify-7_16_2025, 9_05_40 pm.csv")
gdf = gpd.GeoDataFrame(
    df,
    geometry=gpd.points_from_xy(df["lon"], df["lat"]),
    crs="EPSG:4326"
)
```

```python
gdf = gdf.to_crs(epsg=3857)
```

```python
ax = gdf.plot(figsize=(15, 10), alpha=0.5, edgecolor="k")
ctx.add_basemap(ax)
ax.set_axis_off()
```

```python
gdf.to_file("filipino_restaurants.geojson", driver="GeoJSON")
```
